```python

for i in range(len(df)):
    track_uri = df["spotifyid"][i].strip()
    if track_uri[-4:] != "none":
        features = sp.audio_features(track_uri)[0]
        # append the features to the csv file
        df.at[i, "danceability"] = features.get("danceability")
        df.at[i, "energy"] = features.get("energy")
        df.at[i, "key"] = features.get("key")
        df.at[i, "loudness"] = features.get("loudness")
        df.at[i, "mode"] = features.get("mode")
        df.at[i, "speechiness"] = features.get("speechiness")
        df.at[i, "acousticness"] = features.get("acousticness")
        df.at[i, "instrumentalness"] = features.get("instrumentalness")
        df.at[i, "liveness"] = features.get("liveness")
        df.at[i, "valence"] = features.get("valence")
        df.at[i, "tempo"] = features.get("tempo")
        df.at[i, "duration_ms"] = features.get("duration_ms")
        df.at[i, "time_signature"] = features.get("time_signature")
    else:
        df.at[i, "danceability"] = "None"
        df.at[i, "energy"] = "None"
        df.at[i, "key"] = "None"
        df.at[i, "loudness"] = "None"
        df.at[i, "mode"] = "None"
        df.at[i, "speechiness"] = "None"
        df.at[i, "acousticness"] = "None"
        df.at[i, "instrumentalness"] = "None"
        df.at[i, "liveness"] = "None"
        df.at[i, "valence"] = "None"
        df.at[i, "tempo"] = "None"
        df.at[i, "duration_ms"] = "None"
        df.at[i, "time_signature"] = "None"
```

```python
# fill the 4th row, with mean of 5 rows values for each coloumn in only, danceability, energy, key, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration_ms, time_signature
def idx_to_fill(df, idx):
    # replace cells of None with 0
    for key in df.keys():
        if key != "song" and key != "spotifyid":
            for i in range(len(df)):
                if df[key][i] == "None":
                    df.at[i, key] = 0
    for key in df.keys():
        if key != "song" and key != "spotifyid":
            df.at[idx, key] = df[key].mean()

    df.at[idx, "key"] = round(df["key"].mean())
    df.at[idx, "mode"] = round(df["mode"].mean())
    df.at[idx, "time_signature"] = round(df["time_signature"].mean())
    return df

df = idx_to_fill(df, 3)
df = idx_to_fill(df, 8)
```

```python
# make plots for each of the columns individually and show them in notebook with x axis as song name and y axis as coloumn of which the plot is
import matplotlib.pyplot as plt
import numpy as np

def plot(df, key):
    plt.figure(figsize=(10, 5))
    plt.bar(df["song"], df[key], color = 'blue')
    plt.xlabel("Songs")
    plt.ylabel(key)
    plt.xticks(rotation=90)
    plt.title(key + " vs Songs")
    plt.show()

for key in df.keys():
    if key != "song" and key != "spotifyid":
        plot(df, key)
```
